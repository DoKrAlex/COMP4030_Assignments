In this assignment, we will use the following definition of `backtrack`. It is slightly more complicated than the version we studied in class.

To do this assignment, you will have to understand the lecture notes on `backtracking`.  You will need to review those notes first.

Note that in this assignment, we must not modify the definition of this version of `backtrack`.  You need to understand how it works, and how to use it by definining parameters properly.

````python
def backtrack(solution, i, possibilities, is_feasible=lambda sol,i: True,  get_solution=lambda sol: sol):
    if is_feasible(solution, i):
        if i==len(solution):  # everything is configured
            print(get_solution(solution))
        else:
            for p in possibilities:
                solution[i] = p
                backtrack(solution, i+1, possibilities, is_feasible, get_solution)
                
````

How backtracking works:
* Lines 6-8: assign solution[i] to each possible value; and for each value, use the same strategy recursively to generate all solutions with "this configuration".
* When we call `backtrack(solution, 10, possibilities)`, solution is already "configured" from index 0 to 9.
* The result of calling `backtrack(solution, 10, possibilities)` is a generation (printing out) of all solutions with this configuration.

````python
backtrack(
    [None,None,None],  # nothing is configured. So, this generates *all* solutions
    0,                 # we have to start at level 0, if nothing is configured.
    [True,False],      # a set is a list of boolean values
)
````

Create a custom defintion of `get_solution` and `is_feasible`. Then, use `backtrack` to generate all sets containing all sets with the following colors: red, green, blue, orange, yellow.  However, these sets must have the following constraints:
* Each set cannot have both colors `red` and `orange`.
* Each set cannot have both colors `blue` and `yellow`.

The backtracking should print out something like this:

```python
{'red', 'blue', 'green'}
{'red', 'yellow', 'green'}
{'red', 'green'}
{'red', 'blue'}

set()
```


ANSWER:

````python
colors = ['orange', 'blue', 'green', 'red', 'yellow']

def get_set(solution):
    return set([colors[i] for i in range(len(solution)) if solution[i]==True])

def my_criterion(solution, i):
    if i>3:
        if solution[0]==True and solution[3]==True:
            return False
    if i>4:
        if solution[1]==True and solution[4]==True:
            return False
    return True

backtrack(
    [None]*len(colors),
    0,           
    [True,False], 
    get_solution = get_set,
    is_feasible = my_criterion
)
````

**Problem 8**

Continuing from the previous problem, generate all solutions that meet the following constraints:
+ 1900, 3825 are scheduled at TR0940.
+ 4081 is scheduled at TR1440.
+ No more than two courses are scheduled in one time block.



ANSWER:

````python
courses = ['1900', '2150', '3115', '3825', '4030', '4040', '4081']

timeblocks = ['TR0940', 'TR1120', 'TR1300','TR1440']

def scheduling_constraints(solution, i):
    if i>3:
        if solution[0] != 'TR0940' or solution[3] != 'TR0940':
            return False
        
    if i>6:
        if solution[6] != 'TR1440':
            return False
        
    # check to make sure no more than one courses are scheduled in one time block.
    # we will look at solution[i-1], this is the last thing we configured.
    # if the current solution (from index 0 to i-1) violates our constraints, it's the solution[i-1]
    if i>2 and solution[0:i].count(solution[i-1]) > 2:
        return False
    
    return True

backtrack(
    solution = [None]*7,
    i = 0,
    possibilities = timeblocks,
    is_feasible = scheduling_constraints,
)
````
